```python
import cv2
image_path = 'e1.jpeg'
```

```python
img = cv2.imread(image_path)
```

```python
img.shape
```

```python
gray_scale = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
gray_scale.shape
```

```python
face_capture = cv2.CascadeClassifier("C:/Users/hp/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml")
```

```python
face = face_capture.detectMultiScale(gray_scale, scaleFactor= 1.2, minNeighbors = 5, minSize=(5,5))
for (x,y,w,h) in face:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),1)
```

```python
img_rgb1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.imshow(img_rgb1)
plt.axis('off')
```
