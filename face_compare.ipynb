```python
import cv2
```

```python
img1 = cv2.imread('e1.jpeg')
img2 = cv2.imread('e2.jpeg')
```

```python
gry_img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
gry_img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
print(gry_img1.shape)
print(gry_img2.shape)
```

```python
#Detecting Faces
face_detector = cv2.CascadeClassifier("C:/Users/hp/anaconda3/Lib/site-packages/cv2/data/haarcascade_frontalface_default.xml") 
face1 = face_detector.detectMultiScale(gry_img1,1.1,4)
face2 = face_detector.detectMultiScale(gry_img2,1.1,4)
```

```python
for(x,y,w,h) in face1:
    cv2.rectangle(img1, (x,y), (x+w, y+h), (255,0,0), 1)
```

```python
for(x,y,w,h) in face2:
    cv2.rectangle(img2, (x,y), (x+w, y+h), (255,0,0), 1)
```

```python
image1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
fig,ax = plt.subplots(1,2)
ax[0].imshow(image1)
ax[1].imshow(image2)
ax[0].axis('off')
ax[1].axis('off')
if len(face1) == len(face2):
    print("Input Images Matched")
else:
    print("Not Matched")
```
